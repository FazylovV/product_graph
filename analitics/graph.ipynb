```python
import pandas as pd
import networkx as nx
```

```python
df = pd.read_csv('data/pairs_of_products.csv')
df
```

```python
G = nx.DiGraph()
```

```python
for _, row in df.iterrows():
    source_product = row.iloc[0]
    target_product = row.iloc[1]
    G.add_edge(source_product, target_product)
```

```python
in_degree_centralities = dict(nx.in_degree_centrality(G))
in_degree_centralities
```

```python
out_degree_centrality = dict(nx.out_degree_centrality(G))
out_degree_centrality
```

```python
# betweenness_centrality = dict(nx.betweenness_centrality(G))
# betweenness_centrality
```

```python
page_rank = dict(nx.pagerank(G))
page_rank
```

```python
metrics = pd.DataFrame({
    'id': list(in_degree_centralities.keys()),
    'in_degree': list(in_degree_centralities.values()),
    'out_degree': list(out_degree_centrality.values()),
    'page_rank': list(page_rank.values())
})
metrics
```

```python
metrics.to_csv('metrics.csv', index=False)
```

```python
top_in_degree = metrics.sort_values(by='in_degree', ascending=False).head(10)
top_in_degree
```

```python
top_out_degree = metrics.sort_values(by='out_degree', ascending=False).head(10)
top_out_degree
```

```python
top_page_rank = metrics.sort_values(by='page_rank', ascending=False).head(10)
top_page_rank['out_degree']
```

```python
products = pd.read_csv('./data/products.tsv', delimiter='\t')
products.head(3)
```

```python

for vertex in list(top_out_degree['id']):
    edges = df[(df['product1'] == vertex) | (df['product2'] == vertex)]
    edges.to_csv('outdegree/edges_' + str(vertex) + '.csv', index=False)
    # vertexes = list(edges['product1'].unique()) + list(edges['product2'].unique())
    # print(vertex)
    # vertexes = set(map(str, list(edges['product1'].unique()) + list(edges['product2'].unique())))
    vertexes = products[(products['id'].isin(edges['product1'])) | (products['id'].isin(edges['product2']))]
    vertexes.to_csv('outdegree/vertex_' + str(vertex) + '.csv', index=False)
```
